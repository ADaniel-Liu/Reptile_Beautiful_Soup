```python
# 获取页面下租房信息，输入网址，返回lxml 处理后的文本
def get_page(url):
    responce = requests.get(url)
#     解决乱码！！！
    responce.encoding = 'utf-8'
    soup = BeautifulSoup(responce.text, 'lxml')  
    return soup

# 获取页面下租房信息，输入网址，返回一个列表(该函数仅适合 0756fang.com)
def get_links(url):
    get_page(url)
    links_div = soup.find_all('div' , class_ = "house-images")
    links = [div.a.get('href') for div in links_div]
    return links
```

```python
# 将爬取的网页前加上https://www.0756fang.com
url = 'https://www.0756fang.com/rent'
link1 = get_links(url)
link2 = ["httzps://www.0756fang.com" + i for i in link1]
link2
```
